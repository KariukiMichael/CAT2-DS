```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
from google.colab import files
data = pd.read_excel("/content/cat2.xlsx")
feature_to_cluster = 'Age'
l= LabelEncoder()
data['Rate of wear'] = l.fit_transform(data['Rate of wear'])
data['performance'] = l.fit_transform(data['performance'])
X = data[[feature_to_cluster, 'Misalignment']]
s= StandardScaler()
X_std = s.fit_transform(X)
c=3
k= KMeans(n_clusters=c, random_state=0)
data['Cluster'] = k.fit_predict(X_std)
title = f'Clusters of {feature_to_cluster} against Misalignment'
plt.figure(figsize=(8, 6))
for i in range(c):
 cluster_data = data[data['Cluster'] == i]
 plt.scatter(cluster_data['Misalignment'], cluster_data[feature_to_cluster], c=plt.cm.viridis(i /
(c- 1)), label=f'Cluster {i+1}')
plt.xlabel('Misalignment')
plt.ylabel(feature_to_cluster)
plt.title(title, fontsize=16, fontweight='bold')
plt.legend(loc='lower right')
plt.show()
```
